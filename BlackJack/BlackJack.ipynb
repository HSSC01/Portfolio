```python
import random
import numpy as np
from IPython.display import clear_output
```

```python
card_numbers = ['A','K','Q','J','10','9','8','7','6','5','4','3','2']
face_cards = ['A','K','Q','J']
card_suits = ['Spades','Hearts','Clubs','Diamonds']
```

```python
def Blackjack(num_decks):    
    global shoe, player_hand, dealer_hand

    def reset_game():
        shoe.clear()
        for _ in range(num_decks):
            for suit in card_suits:
                for number in card_numbers:
                    shoe.append(f'{suit[0]}{number}')


    def deal_cards():
        player_hand, dealer_hand = [], []
        for _ in range(2):
            player_card = random.choice(shoe)
            shoe.remove(player_card)
            player_hand.append(player_card)

            dealer_card = random.choice(shoe)
            shoe.remove(dealer_card)
            dealer_hand.append(dealer_card)

        return player_hand, dealer_hand

    def get_values(hand):
        value = 0
        aces = 0
        for card in hand:
            rank = card[1:]
            if rank in face_cards:
                if rank == 'A':
                    value += 11
                    aces += 1
                else:
                    value += 10
            else:
                value += int(rank)
        while value > 21 and aces != 0:
            value -= 10
            aces -= 1
        return value

    def player_value():
        return get_values(player_hand)
    
    def dealer_value():
        return get_values(dealer_hand)
    
    def hit_or_stand():
        while True:
            decision = input('Hit or Stand? (h/s): ').lower()
            if decision in ['hit', 'h']:
                new_card = random.choice(shoe)
                shoe.remove(new_card)
                player_hand.append(new_card)
                print(f'You drew {new_card}, total is now {player_value()}')

                if player_value() > 21:
                    print('You bust!')
                    return 'bust'
                elif player_value() == 21:
                    print('You have 21!')
                    break
            elif decision in ['stand','s']:
                print(f'You stood on {player_value()}')
                break
            else:
                print('Invalid choice. Please enter "hit" or "stand".')
        return 'stand'
    
    def check_blackjack():
        if player_value() == 21:
            if dealer_value() == 21:
                print("Both have Blackjack! Push!")
            else:
                print("Blackjack! You win!")
            return True
        if dealer_value() == 21:
            print("Dealer has Blackjack! You lose.")
            return True
        return False
    
    def dealer_turn():
        print(f'Dealer has {dealer_value()}')
        while dealer_value() < 17:
            new_card = random.choice(shoe)
            shoe.remove(new_card)
            dealer_hand.append(new_card)
            print(f'Dealer hits and draws {new_card}, total is now {dealer_value()}')
        
        if dealer_value() > 21:
            print('Dealer busts! You win!')
        elif dealer_value() == player_value():
            print("Push!")
        elif dealer_value() > player_value():
            print(f'Dealer wins with {dealer_value()} against your {player_value()}')
        else:
            print(f'You win with {player_value()} against dealer\'s {dealer_value()}')
    
    def again():
        choice = input('Play again? (y/n)').lower()
        if choice in ['yes','y']:
            clear_output(wait=True)
            play()
        elif choice in ['no','n']:
            print('Thanks for playing!')
            return
        else:
            print("Invalid choice. Please enter 'y' or 'n'.")
            return again()

    def play():
        global player_hand, dealer_hand
        if len(shoe) < 52*num_decks*0.25:
            print("New shoe!")
            answer = input('Continue with new shoe? (y/n)').lower()
            if answer in ['yes','y']:
                reset_game()
                return play()
            elif answer in ['no','n']:
                print('Thanks for playing!')
                return
            else:
                print("Invalid choice. Please enter 'y' or 'n'.")
                return play()
        
        player_hand, dealer_hand = deal_cards()

        print(f'\nYou have {player_hand[0]} and {player_hand[1]}, totalling {player_value()}')
        print(f'Dealer shows {dealer_hand[0]}')

        if check_blackjack():
            return again()

        result = hit_or_stand()

        if result == 'bust':
            return again()

        
        dealer_turn()
        return again()
        
    
    shoe = []
    reset_game()
    play()

    return
```
